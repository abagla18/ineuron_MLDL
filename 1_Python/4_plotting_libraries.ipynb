## Plotting with pandas

```python
df = pd.DataFrame(randn(10,4), index = pd.date_range('1/1/2000', periods = 10), columns=list('ABCD'))
df
df.plot();
```

```python
df['A'].plot()
```

```python
df1 = pd.DataFrame(randn(10,3), index = pd.date_range('1/1/2000', periods = 10), columns=['A','B','C'])
df1
```

```python
df1.plot.bar()
```

```python
df1.plot.hist(bins=20)
```

```python
df2 = pd.DataFrame({'a':np.random.randn(1000)+1, 'b':np.random.randn(1000),'c':np.random.randn(1000)-1},columns =['a','b','c'])
df2.plot.hist(bins=20)
```

```python
df3 = pd.DataFrame(np.random.rand(10,4), columns=['A','B','C','D'])
df3.plot.box()
```

```python
df3.plot.area();
```

```python
df3 = pd.DataFrame(np.random.rand(50,4), columns=['A','B','C','D'])
df3.plot.scatter(x='A',y='B')
```

```python
df3['A'].plot.pie(subplots=True);
```

## Matplotlib

```python
#plt.style.use()
x = np.linspace(0,10,100)
y = 4+2*np.sin(2*x)
# plot
fig,ax = plt.subplots()
print(ax)

ax.plot(x,y,linewidth=2)
ax.set(xlim=(0,8),xticks=np.arange(1,8),
       ylim=(0,8),yticks=np.arange(1,8))

plt.show();
```
