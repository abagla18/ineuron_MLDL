```python
#plt.style.use()
x = np.linspace(0,10,100)
y = 4+2*np.sin(2*x)
# plot
fig,ax = plt.subplots()
print(ax)

ax.plot(x,y,linewidth=2)
ax.set(xlim=(0,8),xticks=np.arange(1,8),
       ylim=(0,8),yticks=np.arange(1,8))

plt.show();
```
